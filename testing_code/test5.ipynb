```python
import os
import json
from dotenv import load_dotenv
from huggingface_hub import InferenceClient
from autogen import AssistantAgent, UserProxyAgent

# Load environment variables from .env file
load_dotenv('/Users/jasper/Downloads/My_first_agents-main/notebooks/.env')

# Verify that the OpenAI API key is loaded
open_ai_api_key = os.getenv('OPENAI_API_KEY')
if not open_ai_api_key:
    raise ValueError("OPENAI_API_KEY not found in environment variables")

# Verify that the Hugging Face API key is loaded
hf_api_key = os.getenv('HUGGINGFACE_API_KEY')
if not hf_api_key:
    raise ValueError("HUGGINGFACE_API_KEY not found in environment variables")

# Define the repository ID for the Hugging Face model
repo_id = "Qwen/Qwen2-VL-7B-Instruct"

# Initialize the InferenceClient with the API key
llm_client = InferenceClient(
    model=repo_id,
    token=hf_api_key,  # Pass the API key
    timeout=120,
)

def call_llm(inference_client: InferenceClient, prompt: str):
    try:
        response = inference_client.post(
            json={
                "inputs": prompt,
                "parameters": {
                    "max_new_tokens": 150,  # Adjusted to limit the length of the response
                    "top_p": 0.95,          # Adjusted for more diverse responses
                    "temperature": 0.8      # Adjusted for more creative responses
                },
                "task": "text-generation",
            },
        )
        response_json = json.loads(response.decode('utf-8'))
        return response_json[0]["generated_text"]
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Configuration for the LLM
config_list = [
    {
        "model": repo_id,  # Use the Hugging Face model
        "client": llm_client,  # Pass the InferenceClient
    }
]

llm_config = {"config_list": config_list}

# Initialize the AssistantAgent
assistant = AssistantAgent("assistant", llm_config=llm_config)

# Initialize the UserProxyAgent
user_proxy = UserProxyAgent(
    "user_proxy",
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith('TERMINATE'),
    human_input_mode='NEVER',
    max_consecutive_auto_reply=5
)

# Define a function to integrate the Hugging Face model call with the Autogen framework
def integrate_hf_model(assistant, user_proxy, prompt):
    generated_text = call_llm(llm_client, prompt)
    if generated_text:
        chat_result = user_proxy.initiate_chat(assistant, message=generated_text)
        return chat_result
    else:
        return "Error in generating text from Hugging Face model."

# Call the function with a prompt
prompt = "Write a shell script to display all files in this system"
chat_result = integrate_hf_model(assistant, user_proxy, prompt)

print(chat_result)
```
